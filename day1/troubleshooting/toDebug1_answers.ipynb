Background information:
- Sheet 1 of the Excel file has coded data about animals found in a zoo. One column is "type" which contains a numerical code from 1-7. 
- Sheet 2 of the Excel file contains the key to what those 7 categories are - classes of animals, such as "mammal" and "bird".
- This code is meant to create a new csv file with the animal data by replacing the "type" column (data is numerical, 1-7) with a new column called "class" (data is text; "mammal", "bird", etc.).

```python
import pandas as pd
```

```python
filename = "zoo.xlsx"
output = "cleanedZoo.csv"
```

```python
def cleanFile(infile, outfile):
    """
    Cleans the zoo dataset by reading an Excel file, 
    mapping animal types to classes based on the key-value pair provided 
    in a second sheet, and saving the cleaned data to a CSV file.
    """
    df = pd.read_excel(infile, sheet_name=0)
    class_df = pd.read_excel(infile, sheet_name=1)

    # the first error comes from trying to map the columns together incorrectly
    # to do this, you need to create a dict and use the map function
    class_dict = dict(zip(class_df.iloc[:, 0], class_df["class"]))
    df["class"] = df["type"].map(class_dict)

    # there was a second error in the drop command: it was missing the axis
    # (otherwise pandas would look for a row named "type")
    df = df.drop("type", axis=1)
    df.to_csv(outfile)
```

```python
cleanFile(filename, output)
```
